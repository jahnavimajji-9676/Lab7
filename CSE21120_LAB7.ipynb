```python
#A1
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm

# Read data from the CSV file
data = pd.read_csv('the_final_no_changes_data.csv')

# Assuming 'polarity', 'subjectivity', 'len' are features and 'category' is the target
X = data[['polarity', 'subjectivity', 'len']].values
y = data['category'].apply(lambda x: 1 if x > 20 else 0).values  # Binary classification based on a threshold

# Splitting dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# SVM classifier
clf = svm.SVC()

# Fit the SVM to the training data
clf.fit(X_train, y_train)

# Get the support vectors
support_vectors = clf.support_vectors_

print("Support Vectors:\n", support_vectors)
```

```python
#A3
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

# Read data from the CSV file
data = pd.read_csv('the_final_no_changes_data.csv')

# Assuming 'polarity', 'subjectivity', 'len' are features and 'category' is the target
X = data[['polarity', 'subjectivity', 'len']].values
y = data['category'].apply(lambda x: 1 if x > 20 else 0).values  # Binary classification based on a threshold

# Splitting dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# SVM classifier
clf = svm.SVC()

# Fit the SVM to the training data
clf.fit(X_train, y_train)

# Use the predict function 
predicted_values = clf.predict(X_test)
print("Predicted values:", predicted_values)

# Relate the output values to the class values predicted
for i in range(len(X_test)):
    custom_accuracy = sum(predicted_values == y_test) / len(y_test)
    print("Custom Accuracy:", custom_accuracy)
```

```python
#A4
from sklearn.metrics import accuracy_score
kernel_functions = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernel_functions:
    # Creating an SVM classifier with the current kernel function
    clf = svm.SVC(kernel=kernel)
    # Fit the classifier to the training data
    clf.fit(X_train, y_train)
    # Use the classifier to make predictions on the test set
    y_pred = clf.predict(X_test)
    # Calculating and print accuracy for the current kernel
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy with'{kernel}' kernel: {accuracy:.2f}")
```
